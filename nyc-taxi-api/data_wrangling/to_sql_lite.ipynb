```python
import pandas as pd
import sqlite3
import os

def parquet_to_sqlite(parquet_folder, sqlite_db_path, table_name):
    # Crear o conectar a la base de datos SQLite
    conn = sqlite3.connect(sqlite_db_path)
    
    # Leer todos los archivos Parquet en la carpeta
    for file in os.listdir(parquet_folder):
        if file.endswith(".parquet"):
            file_path = os.path.join(parquet_folder, file)
            # Leer el archivo Parquet en un DataFrame
            df = pd.read_parquet(file_path)
            # Escribir el DataFrame en la tabla de SQLite
            df.sample(10000).to_sql(table_name, conn, if_exists="append", index=False)
            print(f"Archivo {file} cargado exitosamente.")

    # Cerrar la conexión a la base de datos
    conn.close()
```

```python

# Ejemplo de uso
parquet_folder = "./parquet_data"
sqlite_db_path = "nyc_taxi.db"
table_name = "records"

parquet_to_sqlite(parquet_folder, sqlite_db_path, table_name)
```

```python
# Conectar a la base de datos SQLite
conn = sqlite3.connect('nyc_taxi.db')

# Leer todos los datos de la tabla 'records' y cargarlos en un DataFrame
df = pd.read_sql_query("SELECT * FROM records", conn)

# Cerrar la conexión a la base de datos
conn.close()
```
